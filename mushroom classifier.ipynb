# Feature Engineering

- converting categorical variables into numeric form using label encoder
- using label encoder because we can see that in dataset we have ordinal categories for each and every features

```python
#dropping irrelevent colunm
df = df.drop(["veil-type"],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
for i in df.columns:
    df[i]=label.fit_transform(df[i])
```

```python
df.head()
```

```python
df.info()
```

## Feature Selection

- using mutual information to select top 10 features for model building

```python
X = df.copy()
y = X.pop('class')
```

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
#creating object mic
#using scoring function as mutual_info_classif
# k=10 for selcting top 10 values
mic = SelectKBest(score_func= mutual_info_classif, k= 10)
```

```python
#getting mutual information score
mic.fit (X,y)
MI_score = pd.Series(mic.scores_, index =X.columns)
MI_score.sort_values(ascending = False)
```

```python
sns.barplot(x = MI_score, y=MI_score.index)
```

```python
#getting top 10 high score features
df=mic.fit_transform(X,y)
df.shape
```

```python
df
```

```python
df= pd.DataFrame(df)
```

```python
df.head()
```

# Model Buliding

- Logistic Regression
- Decison Tree
- Random Forest

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.2, shuffle=True , random_state=0)
```

```python
 X_train.shape
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logmodel= LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
predections= logmodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predections)
```
